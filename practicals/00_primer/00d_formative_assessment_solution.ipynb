## 2. Visualise data

üñåÔ∏è Plot histograms of three variables of your choice that contain interval/ratio measurements. Implement the following settings:
- 12 bins per histogram
- trait name as x axis label
- grid lines behind bars

```python
import matplotlib.pyplot as plt


# example variables chosen:
vars_subset = ['5-1_AdultBodyMass_g', '9-1_GestationLen_d', '26-1_GR_Area_km2']

for var in vars_subset:
    plt.figure()
    plt.hist(data[var].dropna(), bins=12)
    plt.xlabel(var)
    plt.grid()
    plt.show()
```

üñåÔ∏è Create a scatter plot of litter size (x) _vs._ neonate body mass (x) with the following properties:
- x and y axis labels
- grid in the background

```python
plt.figure()

plt.scatter(data['15-1_LitterSize'],
            data['5-3_NeonateBodyMass_g'])
ax = plt.gca()
ax.set_yscale('log')
plt.xlabel('Litter size')
plt.ylabel('Log neonate body mass [g]')
plt.grid()
plt.show()
```

This plot doesn't really tell us much yet, possibly because the body mass is in grams and therefore
distributed across a very large range.

As you have seen in previous modules, we can improve our visualisation here if we show the log of
that variable instead.

üñåÔ∏è Change the y axis in the code cell above to log scale. You can either do this in Matplotlib (use
a Web search engine to figure it out) or pre-log-transform the y axis variable.

üñåÔ∏è Create the same plot again below, but only for the following orders:
- _Carnivora_
- _Chiroptera_
- _Rodentia_

Colour them all differently, with labels and legend.

üí° The easiest way to do so is with a `for` loop over all possible values. To get a legend, you can
add an input value `label=...` when calling the `plt.scatter(...)` function, and `plt.legend()` at
the very end outside the loop.

```python
plt.figure()

for order in ('Carnivora', 'Chiroptera', 'Rodentia'):
    valid = data['MSW05_Order'] == order
    plt.scatter(data['15-1_LitterSize'][valid],
                data['5-3_NeonateBodyMass_g'][valid],
                label=order)
ax = plt.gca()
ax.set_yscale('log')
plt.xlabel('Litter size')
plt.ylabel('Log neonate body mass [g]')
plt.grid()
plt.legend()
plt.show()
```

## 3. Basic correlations

üñåÔ∏è Find out how much variables that contain floating point values correlate with each other. Visualise the outcome on a grid, choosing a [diverging colour map](https://matplotlib.org/stable/users/explain/colors/colormaps.html#diverging).

üí° Tips:
1. Look up the [select_dtypes](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.select_dtypes.html) function.
2. The [corr](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html) function will also come in handy.
3. `plt.imshow()` will be your friend for the visualisation.
4. You may want to increase the figure size (`plt.figure(figsize=(14,14))` and to set x and y [ticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html)).

```python
float_cols = data.select_dtypes('float64')
corr = float_cols.corr()

# visualise
plt.figure(figsize=(14,14))
plt.imshow(corr, cmap='BrBG')
plt.xticks(range(len(corr)), float_cols.columns, rotation=90)
plt.yticks(range(len(corr)), float_cols.columns)
plt.colorbar()
plt.show()
```
